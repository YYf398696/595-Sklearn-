```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
import pandas as pd
np.set_printoptions(suppress=True)
```

```python
iris=pd.read_csv('iris.csv')
wine=pd.read_csv('wine_data.csv')
```

```python
iris.columns=["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width","Type"]
```

```python
K = range(1, 10)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(iris)
    meandistortions.append(sum(np.min(cdist(iris, kmeans.cluster_centers_, 'euclidean'), axis=1)) /iris.shape[0])
```

```python
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method',fontsize=20)
plt.show()
```

We can see that the elbow is at k=3 for the iris dataset

```python
K = range(1, 10)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(wine)
    meandistortions.append(sum(np.min(cdist(wine, kmeans.cluster_centers_, 'euclidean'), axis=1)) /wine.shape[0])
```

```python
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method',fontsize=20)
plt.show()
```
